```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Ecommerce Project") \
    .getOrCreate()
```

```python
df = spark.read.csv("/content/online_retail.csv", header=True, inferSchema=True)
df.show(5)
```

```python
print("✅ Schema:")
df.printSchema()
```

```python
print("✅ First 5 Rows:")
df.show(5)
```

```python
df = df.dropna()
df = df.dropDuplicates()

print("✅ After Cleaning (Null & Duplicates Removed):")
df.show(5)
```

```python
df = df.filter(df.Quantity > 0)

print("✅ After Removing Negative Quantity:")
df.show(5)
```

```python
from pyspark.sql.functions import col

df = df.withColumn("TotalPrice", col("Quantity") * col("UnitPrice"))

print("✅ TotalPrice Column Added:")
df.show(5)
```

```python
print("✅ Total Revenue:")
df.selectExpr("sum(TotalPrice) as TotalRevenue").show()
```

```python
print("✅ Country-wise Sales:")
df.groupBy("Country").sum("TotalPrice").show()
```

```python
print("✅ Top 10 Products:")
df.groupBy("Description") \
  .sum("TotalPrice") \
  .orderBy("sum(TotalPrice)", ascending=False) \
  .show(10)
```

```python
print("✅ Top 10 Customers:")
df.groupBy("CustomerID") \
  .sum("TotalPrice") \
  .orderBy("sum(TotalPrice)", ascending=False) \
  .show(10)
```

```python
df.toPandas().to_csv("cleaned_data.csv", index=False)

print("✅ File saved as cleaned_data.csv")
```
